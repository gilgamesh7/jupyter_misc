#### API - COVID

```python
import requests
from datetime import date,datetime,timedelta

today = date.today()
yesterday = today - timedelta(days=7)

country = "india"
endpoint = f"https://api.covid19api.com/country/{country}/status/confirmed"
params = {"from":str(yesterday), "to":str(today)}

response=requests.get(endpoint,params=params).json()

total_cases = 0
for day in response :
    print(f"{day['Date']} - {day['Cases']}")
```

#### Scrape links using bs4

```python
import bs4
import requests

resp = requests.get("https://pybit.es/archives")
# print(resp.text)
soup=bs4.BeautifulSoup(resp.text,'html.parser')
# print(soup)
# Get only links with text 'class' in them
# articles=[link['href'] for link in soup.find_all('a',href=True) if 'class' in link.text.lower()]
# Get links with https as part of the link
articles=[link['href'] for link in soup.find_all('a',href=True) if 'https' in link['href']]
for article in articles:
    print(article)
```

#### Python to log into a PHP website
##### https://linuxhint.com/logging_into_websites_python/

```python
from requests import Session
from bs4 import BeautifulSoup as bs
 
with Session() as s:
    site = s.get("http://quotes.toscrape.com/login")
    bs_content = bs(site.content, "html.parser")
    token = bs_content.find("input", {"name":"csrf_token"})["value"]
    login_data = {"username":"admin","password":"12345", "csrf_token":token}
    print(login_data)
    s.post("http://quotes.toscrape.com/login",login_data)
    home_page = s.get("http://quotes.toscrape.com")
    print(home_page.content)
```

### Element Tree from XML

```python
import xml.etree.ElementTree as ET
from collections import Counter
import requests

response = requests.get("https://pybit.es/feeds/all.rss.xml")

tree=ET.fromstring(response.content)
print(tree)

titles = [e.text for e in tree.findall("channel/item/title")]
for title in titles:
    print(title)
```
