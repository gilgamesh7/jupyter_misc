```python
crimes_small_mi['Occurances'] = np.ones(len(crimes_small_mi),dtype=int)
```

```python
crimes_grouped = crimes_small_mi.Occurances.groupby(level=[0,1]).sum()
```

```python
crimes_grouped
```

```python
crimes_by_type=crimes_grouped.unstack(level=1).fillna(0).resample('h').sum().astype(int)
crimes_by_type
```

```python
crimes_by_type.loc[:,['THEFT','ASSAULT','BURGLARY']].plot()
```
